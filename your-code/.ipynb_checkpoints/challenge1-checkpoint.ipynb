```python
x_values = [17,51,27,21,36,48,19,26,54,30]
y_values = [10,1,5,9,4,2,9,6,0,3]

student_data = pd.DataFrame({'X': x_values, 'Y': y_values})

student_data
```

Do you see a trend? Can you make any hypotheses about the relationship between age and number of tardies?

Calculate the covariance and correlation of the variables in your plot. What is the difference between these two measures? Compare their values. What do they tell you in this case? Add your responses as comments after your code.

```python
# Your response here.
#correlation:
corr_values = np.corrcoef(x_values, y_values)
corr_values
```

```python
#covariance:
cov_values = np.cov(x_values,y_values)
cov_values
```

```python
# Negative and strong correlation in both measures (corr and cov), showed in (1,2) and (2,1)
```

Build a regression model for this data. What will be your outcome variable? What type of regression are you using? Add your responses as comments after your code.

```python
x_values
```

```python
x_values = np.array(x_values)
y_values = np.array(y_values)

x_values = x_values.reshape(-1, 1) 

model = linear_model.LinearRegression()
result = model.fit(x_values,y_values)

intercept = result.intercept_
coef = result.coef_
regression_line = result.intercept_ + result.coef_[0]*x_values
print(f'coef= {coef}, intercept= {intercept}')
```

## Challenge 2: Error Analysis

I am suspicious about the last few parties I have thrown: it seems that the more people I invite the more people are unable to attend. To know if my hunch is supported by data, I have decided to do an analysis. I have collected my data in the table below, where X is the number of people I invited, and Y is the number of people who attended. 

|  X |  Y |
|----|----|
| 1  |  1 |
| 3  |  2 |
| 4  |  4 |
| 6  |  4 |
| 8  |  5 |
| 9  |  7 |
| 11 |  8 |
| 14 |  13 |

We want to know if the relationship modeled by the two random variables is linear or not, and therefore if it is appropriate to model it with a linear regression. 
First, build a dataframe with the data. 

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
attendances = pd.DataFrame({'X':[1,3,4,6,8,9,11,14],'Y':[1,2,4,4,5,7,8,13]})
attendances
```

Draw a dispersion diagram (scatter plot) for the data, and fit a regression line.

```python
X = np.array(attendances['X'])
Y = np.array(attendances['Y'])

X = X.reshape(-1, 1) 

model = linear_model.LinearRegression()
result = model.fit(X,Y)

intercept = result.intercept_
coef = result.coef_
regression_line = result.intercept_ + result.coef_[0]*X

plt.scatter(X,Y)
plt.plot(X, regression_line, c = 'orange')

plt.show()
```

What do you see? What does this plot tell you about the likely relationship between the variables? Print the results from your regression.

```python
print("Intercept:", result.intercept_)
print("Coefficients:", result.coef_)
r_squared = result.score(X, Y) 
print("R-squared:", r_squared)
```

```python
'''The value of R-squared (0.9316696375519905) provides information about the goodness of fit of the model. 
In this case, it indicates that approximately 93.17% of the variability in the dependent variable (Y)
can be explained by the independent variable (X) in the regression model. 
This suggests a strong linear relationship between the two variables and that the model can make good predictions.'''
```

Do you see any problematic points, or outliers, in your data? Remove these points and recalculate your regression. Print the new dispersion diagram with your new model and the results of your model. 

```python
# To remove outliers I have going to use the Interquartile Range (IQR) method, using NumPy

Q1 = np.percentile(Y, 25) 
Q3 = np.percentile(Y, 75) 
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Y_filtered = Y[(Y >= lower_bound) & (Y <= upper_bound)]
```

```python
X = np.array(attendances['X'])
Y_filtered = np.array(attendances['Y'])

X = X.reshape(-1, 1) 

model = linear_model.LinearRegression()
result = model.fit(X,Y_filtered)

intercept = result.intercept_
coef = result.coef_
regression_line = result.intercept_ + result.coef_[0]*X

plt.scatter(X,Y_filtered)
plt.plot(X, regression_line, c = 'orange')

plt.show()
```

```python
# It didn't work, so I remove that value manually

attendances2 = pd.DataFrame({'X':[1,3,4,6,8,9,11],'Y':[1,2,4,4,5,7,8]})
X2 = np.array(attendances2['X'])
Y2 = np.array(attendances2['Y'])

X2 = X2.reshape(-1, 1) 

model = linear_model.LinearRegression()
result = model.fit(X2,Y2)

intercept = result.intercept_
coef = result.coef_
regression_line = result.intercept_ + result.coef_[0]*X2

plt.scatter(X2,Y2)
plt.plot(X2, regression_line, c = 'orange')

plt.show()
```

What changed? Based on the results of the two models and your graphs, what can you say about the form of the data with the problematic point and without it?

```python
'''The coefficient has decreased, resulting in a shallower slope for the linear regression. 
Removing the outlier has significantly improved the clarity of the trend without letting specific outliers 
heavily influence the overall view.'''
```
